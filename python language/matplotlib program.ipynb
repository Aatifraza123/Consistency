```python
#for ploting simple graph
import matplotlib.pyplot as plt
import numpy as np


xpts = np.array([5, 10, 15, 20, 25])
ypts = np.array([3, 6, 9, 12, 15])

plt.plot(xpts, ypts, color='red')
plt.show()
```

#basic graph

```python
xpts = np.array([5, 10, 15, 20, 25])
ypts = np.array([3, 6, 9, 12, 15])

plt.plot(xpts, ypts, color='red',marker='*',linestyle='--', markeredgecolor='blue')
plt.show()
```

```python
a = [2, 4, 6, 8]
b = [3, 6, 9, 12]
plt.plot(a, b, 'r*--')
plt.show()
```

```python
#for heading of the graph
a = [2, 4, 6, 8]
b = [3, 6, 9, 12]
plt.plot(a, b, 'g*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['single element'])
plt.savefig('firstgraph',dpi=200)
plt.title('firstgraph',fontdict={'fontname':'comic sans MS','fontsize':16})
plt.show()
```

```python
x = [2, 4, 6, 8, 10]
y = [3, 6, 9, 12, 15]
plt.plot(x , y, '*--')
plt.legend(["single element"])
plt.xlabel("x-axis")
plt.ylabel("y_axis")
plt.show()
```

```python
y = np.array([40,86,78,90])
name = ['Aatif','Raza','Aamir','Raaz']
nameexplode = [0,0,0,0.2]
plt.pie(y,labels=name,explode=nameexplode,shadow=True)
plt.title('Piegraph',fontdict={'fontname':'comic sans MS','fontsize':16})
plt.savefig('pie.png',dpi=200)
plt.legend()
plt.show()

```

```python
bunch = [0,20,40,60,80,100]
x = [10,20,30,40,50,60]
plt.hist(x,bins=bunch,color='m')
plt.xlabel('bunch')
plt.ylabel('data')
plt.legend(['ok'])
plt.savefig('hist.png',dpi=200)
plt.title('Histgraph',fontdict={'fontname':'comic sans MS','fontsize':16})
plt.show()
```

```python
test = np.random.normal(200,50,400)
test2 = np.random.normal(100,50,300)
test3 = np.random.normal(150,50,350)
test4 = np.random.normal(250,50,450)
data = [test,test2,test3,test4]
plt.boxplot(data)
plt.savefig('box.png',dpi=200)
plt.title('Boxgraph',fontdict={'fontname':'comic sans MS','fontsize':16})
plt.show()
```
